urlretrieve를 사용하여 이미지 및 html파일 저장

```python
# 181213
import sys
import io
import urllib.request
import os # 현재 경로
```

```python
# 원하는 이미지 불러오기(인터넷에서 해당 이미지 주소 복사)
imgUrl ="https://search.pstatic.net/common/?src=http%3A%2F%2Fblogfiles.naver.net%2F20150813_182%2Fdhscjs1785_1439432002539dC72O_JPEG%2FKakaoTalk_20150812_114642594.jpg&type=b400"
htmlURL ="https://www.naver.com/"
```

```python
cwd = os.getcwd() # get the current path

# 저장 경로 설정
savePath1 = cwd + "/download_file/Korean_flag.jpg"
savePath2 = cwd + "/download_file/naver.html"

# urllib.request.urlretrieve(url, 저장경로) : 다운로드하기
urllib.request.urlretrieve(imgUrl, savePath1)
urllib.request .urlretrieve(htmlURL, savePath2)

print("다운로드 완료!")
```
